```python
Amazon["discounted_price"] = Amazon["discounted_price"].replace("[₹,]","",regex=True)
Amazon["discounted_price"] = Amazon["discounted_price"].astype(float)
Amazon["discounted_price"].head() 
```

```python
Amazon["actual_price"] = Amazon["actual_price"].replace("[₹,]","",regex=True)
Amazon["actual_price"] = Amazon["actual_price"].astype(float)
Amazon["actual_price"].head() 
```

```python
Amazon["discount_percentage"] = Amazon["discount_percentage"].replace("%","",regex=True)
Amazon["discount_percentage"] = Amazon["discount_percentage"].astype(float)
Amazon["discount_percentage"].head() 
```

```python
Amazon["rating"] = Amazon["rating"].replace("|","",regex=True)
```

```python
Amazon["rating"] = pd.to_numeric(Amazon["rating"], errors="coerce")
```

```python
Amazon["rating"] = Amazon["rating"].fillna(Amazon["rating"].mean())
```

```python
Amazon["rating"].info()
```

```python
Amazon["rating_count"] = Amazon["rating_count"].replace(",","",regex=True)
Amazon["rating_count"] = Amazon["rating_count"].astype(float)
Amazon["rating_count"].head()
```

```python
Amazon.info()
```

```python
Amazon["rating_count"]=Amazon["rating_count"].fillna(Amazon["rating_count"].mean())
```

```python
Amazon
```

```python
Amazon.info()
```

Amazon.describe()

```python
Amazon["main_category"]=Amazon["category"].str.split("|").str[0]
```

```python
Amazon.info()
```

```python
Amazon["sub_category"] = Amazon["category"].str.split("|").str[1]
Amazon["sub_sub_category"] = Amazon["category"].str.split("|").str[2]
```

```python
Amazon.info()
```

```python
Amazon.head()
```

```python
Amazon = Amazon.drop(columns=["category","about_product","user_id","user_name","review_id","review_title","review_content","img_link","product_link"])
```

```python
value_products = Amazon[(Amazon['rating'] >= 4.0) & (Amazon['discounted_price'] <= 1000)]
print(f"7. Number of high-rated affordable products (≤₹1000): {len(value_products)}")
value_unique = value_products["product_name"].nunique()
value_unique
```

```python
unique_products =Amazon["product_name"].nunique()
print("Total unique products:", unique_products)
```

```python
value_products_percentage = (value_unique / unique_products) * 100
print("Value Product%",value_products_percentage)
```

```python
category_counts = Amazon["main_category"].value_counts()
#plot
plt.Figure(figsize=(10,6))
plt.bar(category_counts.index,category_counts.values,color="skyblue",edgecolor="black")
#add labels and titles
plt.title("Top Categories By Products",fontsize=14)
plt.xlabel("Product Category",fontsize=12)
plt.ylabel("NO.products")
plt.xticks(rotation=45,ha="right")
```

```python
plt.Figure(figsize=(8,5))
plt.hist(Amazon["rating"],bins=10,color="skyblue",edgecolor="black")
#add titles and labels
plt.title("Distribution Of Product Ratings",fontsize=14)
plt.xlabel("Ratings",fontsize=12)
plt.ylabel("Number Of Products",fontsize=12)
plt.grid(axis="y",linestyle="--")
```

```python
plt.Figure(figsize=(8,5))
plt.hist(Amazon["actual_price"],bins=10,color="skyblue",edgecolor="black")
#add titles and labels
plt.title("Distribution Of Product Prices",fontsize=14)
plt.xlabel("prices",fontsize=12)
plt.ylabel("Number Of Products",fontsize=12)
plt.grid(axis="y",linestyle="--")
```

```python
avg_discount = Amazon.groupby("main_category")["discount_percentage"].mean().sort_values(ascending=False)
avg_discount
```

```python
# Remove duplicates based on product_name
Amazon_unique = Amazon.drop_duplicates(subset=["product_name"])

# Now get top 10 most discounted unique products
top_discount = Amazon_unique.sort_values(by="discount_percentage", ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_discount["product_name"], top_discount["discount_percentage"],
         color="blue", edgecolor="black")
plt.title("Most Discounted Products ", fontsize=14)
plt.xlabel("Discount (%)",fontsize=12)
plt.ylabel("Products",fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
